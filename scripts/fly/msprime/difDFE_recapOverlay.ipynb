```python
import tskit, msprime, pyslim
import numpy as np
import gzip
import time
```

```python
# Open runtime file header
#RTfile = open("C:\\Users\\tferrari\\Desktop\\SlimBenchmark\\fly\\data\\msprime_compStat_table.txt", "a")  # append mode

for c in [1000, 500, 100]:
    mu=2.1818e-9*c # Update parameters to scale
    N=652700/c
    rr=2.06e-8*c

    for G in ["1e5", "1e6", "1e7"]:
        x=51 if G=="1e5" else 6 # for 50 or 5 replicates, range is exclusive
        
        for h in ["0.5"]:
            for n in [1,2]:
                for r in range(1,x):
                    start = time.time()
                    
                    # Load in sim tree
                    orig_ts = tskit.load(f'C:\\Users\\tferrari\\Desktop\\SlimBenchmark\\fly\\out\\difDFE\\flyBench_recap_c{c}_{G}_h{h}_n{n}_rep{r}_lowS.trees')
                    print(f"Number of sites: {orig_ts.num_sites}\n"
                          f"Time: {orig_ts.max_root_time}\n"
                          f"Number of mutations: {orig_ts.num_mutations}")
                    
                    # Recapitate tree
                    rts = pyslim.recapitate(orig_ts,
                                            recombination_rate=rr,
                                            ancestral_Ne=N)
                    assert(max([t.num_roots for t in rts.trees()]) == 1)
                    
                    next = pyslim.next_slim_mutation_id(rts)
                    
                    # Overlay neutral mutations
                    ts = msprime.sim_mutations(
                        rts,
                        rate=mu,
                        model=msprime.SLiMMutationModel(type=1,next_id=next),
                        keep=True)
                    
                    print(f"Number of sites: {ts.num_sites}\n"
                          f"Number of mutations: {ts.num_mutations}\n"
                          f"Number of nodes: {ts.num_nodes}")

                    sts = ts.simplify(filter_populations=False,filter_individuals=False,filter_sites=False)
                    
                    # Output recapitated and overlaid tree
                    sts.dump(f'C:\\Users\\tferrari\\Desktop\\SlimBenchmark\\fly\\out\\difDFE\\flyBench_recap_c{c}_{G}_h{h}_n{n}_rep{r}_lowS.over.trees')
                    
                    # Output runtime to file
                    rt = time.time()-start
                    #RTfile.write(f'Recap\t{c}\t{G}\t{h}\t{n}\t{r}\t{rt}\n')
#RTfile.close()
```

```python
# Open runtime file header
#RTfile = open("C:\\Users\\tferrari\\Desktop\\SlimBenchmark\\fly\\data\\msprime_compStat_table.txt", "a")  # append mode

for c in [1000, 500, 100]:
    # mu=2.1818e-9*c # use rate map instead
    N=652700/c
    rr=2.06e-8*c

    for G in ["1e5", "1e6", "1e7"]:
        # Rate map instead of mu
        positions = list(range(4000,int(float(G)),5000)) + list(range(5000,int(float(G)),5000)) + [0,int(float(G))]
        positions.sort()
        rates = list([8.4e-9*c,2.1818e-9*c]*int(float(G)/5000))
        ratemap = msprime.RateMap(position=positions,rate=rates)
        
        x=51 if G=="1e5" else 6 # for 50 or 5 replicates, range is exclusive
        
        for h in ["0.5"]:
            for n in [1,2]:
                for r in range(1,x):
                    start = time.time()
                    
                    # Load in sim tree
                    orig_ts = tskit.load(f'C:\\Users\\tferrari\\Desktop\\SlimBenchmark\\fly\\out\\difDFE\\flyBench_recap_c{c}_{G}_h{h}_n{n}_rep{r}_Swins.trees')
                    print(f"Number of sites: {orig_ts.num_sites}\n"
                          f"Time: {orig_ts.max_root_time}\n"
                          f"Number of mutations: {orig_ts.num_mutations}")
                    
                    # Recapitate tree
                    rts = pyslim.recapitate(orig_ts,
                                            recombination_rate=rr,
                                            ancestral_Ne=N)
                    assert(max([t.num_roots for t in rts.trees()]) == 1)
                    
                    next = pyslim.next_slim_mutation_id(rts)
                    
                    # Overlay neutral mutations
                    ts = msprime.sim_mutations(
                        rts,
                        rate=ratemap,
                        model=msprime.SLiMMutationModel(type=1,next_id=next),
                        keep=True)
                    
                    print(f"Number of sites: {ts.num_sites}\n"
                          f"Number of mutations: {ts.num_mutations}\n"
                          f"Number of nodes: {ts.num_nodes}")

                    sts = ts.simplify(filter_populations=False,filter_individuals=False,filter_sites=False)
                    
                    # Output recapitated and overlaid tree
                    sts.dump(f'C:\\Users\\tferrari\\Desktop\\SlimBenchmark\\fly\\out\\difDFE\\flyBench_recap_c{c}_{G}_h{h}_n{n}_rep{r}_Swins.over.trees')
                    
                    # Output runtime to file
                    rt = time.time()-start
                    #RTfile.write(f'Recap\t{c}\t{G}\t{h}\t{n}\t{r}\t{rt}\n')
#RTfile.close()
```
