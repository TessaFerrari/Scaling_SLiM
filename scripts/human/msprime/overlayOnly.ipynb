```python
import tskit, msprime, pyslim
import numpy as np
import gzip
import time
```

```python
# Create runtime file header
RTfile = open("C:\\Users\\tferrari\\Desktop\\SlimBenchmark\\human\\data\\msprime_compStat_table.txt", "a")  # append mode
RTfile.write("BurnInType\tScalingFactor\tGenomeSize\tDomCoefficent\tBurnNum\tSimNum\tTimeSecs\n")

for c in [10, 5, 1]:
    mu=2.124e-8*c # Update parameter to scale

    for G in ["1e5", "1e6", "1e7"]:
        x=51 if G=="1e5" else 6 # for 50 or 5 replicates, range is exclusive
        
        for h in ["0.0","0.5"]:
            for n in [1,2]:
                for r in range(1,x):
                    start = time.time()
                    
                    # load in sim tree
                    orig_ts = tskit.load(f'C:\\Users\\tferrari\\Desktop\\SlimBenchmark\\human\\out\\toBeOverlaid\\coal\\humanBench_coal_c{c}_{G}_h{h}_n{n}_rep{r}.trees')
                    print(f"Number of sites: {orig_ts.num_sites}\n"
                          f"Time: {orig_ts.max_root_time}\n"
                          f"Number of mutations: {orig_ts.num_mutations}")
                    
                    next = pyslim.next_slim_mutation_id(orig_ts)
                    
                    # Overlay neutral mutations
                    ts = msprime.sim_mutations(
                        orig_ts,
                        rate=mu,
                        model=msprime.SLiMMutationModel(type=1,next_id=next),
                        keep=True)
                    
                    print(f"Number of sites: {ts.num_sites}\n"
                          f"Number of mutations: {ts.num_mutations}\n"
                          f"Number of nodes: {ts.num_nodes}")

                    sts = ts.simplify(filter_populations=False,filter_individuals=False,filter_sites=False)
                    
                    # Output overlaid tree
                    sts.dump(f'C:\\Users\\tferrari\\Desktop\\SlimBenchmark\\human\\out\\msprimetrash\\humanBench_coal_c{c}_{G}_h{h}_n{n}_rep{r}.over.trees')
                    
                    # Output runtime to file
                    rt = time.time()-start
                    RTfile.write(f'Coal\t{c}\t{G}\t{h}\t{n}\t{r}\t{rt}\n')
RTfile.close()
```
